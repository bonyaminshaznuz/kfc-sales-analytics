```python
from sqlalchemy import create_engine

# Load CSV
df = pd.read_csv('KFC_Past_Sales_Dataset.csv')

# Connect to MySQL
engine = create_engine("mysql+pymysql://root:@localhost:3306/KFC_Past_Sales")

# Write
df.to_sql('kfc_sales', con=engine, if_exists='replace', index=False)

# Read
df_mysql = pd.read_sql('kfc_sales', con=engine)
df_mysql.head()
```
